# Importing Libraries

```python
import pandas as pd
import nltk
import re
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from tqdm import tqdm
import time
```

```python
ps = PorterStemmer()
wl = WordNetLemmatizer()
```

```python
# Approx = 3 mins
def Stemming(data):
    new_data_S = []
    start_time_S = time.time()

    for index, row in tqdm(data.iterrows(),total=len(data), desc="Processing"):
        corpus = []
        for col in data.columns:
            if col not in['Date','Label']:
                sent = nltk.sent_tokenize(str(row[col]))
                for i in range(len(sent)):
                    r = re.sub('[^a-zA-Z]', ' ',sent[i])
                    r = r.lower()
                    words = r.split()
                    r = [ps.stem(word) for word in words if word not in set(stopwords.words('english'))]
                    r = ' '.join(r)
                    corpus.append(r)

        new_data_S.append(' '.join(corpus))

    #     new_row_S = {
    #         'Index': index,
    #         'Date': row['Date'],
    #         'Stemmed News': ' '.join(corpus),
    #         'Label': row['Label']
    #     }
    #     new_data_S.append(new_row_S)

    # new_df_S = pd.DataFrame(new_data_S)

    end_time_S = time.time()
    total = end_time_S - start_time_S
    print(f"Total time taken by Stemming = {total : .6f} seconds")
    return new_data_S
```

```python
def Lemmatization(data):
    new_data_L = []

    start_time_L = time.time()

    for index, row in tqdm(data.iterrows(),total=len(data), desc="Processing"):
        corpus = []
        for col in data.columns:
            if col not in['Date','Label']:
                sent = nltk.sent_tokenize(str(row[col]))
                for i in range(len(sent)):
                    r = re.sub('[^a-zA-Z]', ' ',sent[i])
                    r = r.lower()
                    words = r.split()
                    r = [wl.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
                    r = ' '.join(r)
                    corpus.append(r)

        new_data_L.append(' '.join(corpus))

    #     new_row_L = {
    #         'Index': index,
    #         'Date': row['Date'],
    #         'Stemmed News': ' '.join(corpus),
    #         'Label': row['Label']
    #     }
    #     new_data_L.append(new_row_L)

    # new_df_L = pd.DataFrame(new_data_L)

    end_time_L = time.time()
    total = end_time_L - start_time_L
    print(f"Total time taken by Lemmatization = {total : .6f} seconds")
    
    return new_data_L
```
